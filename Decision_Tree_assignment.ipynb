Que1. What is a Decision Tree, and how does it work.

Ans:A Decision Tree is a type of supervised machine learning algorithm used for both classification and regression tasks. It resembles a flowchart-like structure where each internal node represents a decision based on a feature, each branch represents the outcome of the decision, and each leaf node represents a final output (class label or predicted value).

ðŸŒ³ How a Decision Tree Works:

1. Start at the Root Node:

The tree begins at the root, which contains the entire dataset.

2. Split the Data:

At each node, the algorithm chooses the best feature and threshold to split the data. This is done to maximize the separation between classes or minimize prediction error.

Common criteria:

Gini Impurity (used in classification)

Entropy (for information gain)

Mean Squared Error (used in regression)

3. Create Branches:

The data is split into subsets based on the decision rule.

This process repeats recursively for each subset.

4. Stop When...:

A stopping condition is met:

All data in a node belongs to the same class.

Maximum tree depth is reached.

Minimum number of samples per node is too small.

Further splitting doesn't improve the outcome significantly.

5. Prediction:

For a new data point, the tree is traversed from the root to a leaf by following the decision rules. The output at the leaf is the predicted value or class.

Ans: The impurity measures in decisssion tree are

1. Gini Impurity (used in CART - Classification and Regression Trees)

Measures the probability of incorrectly classifying a randomly chosen element if it were randomly labeled according to the distribution in the node.

2.  Entropy (used in ID3 algorithm)
Derived from information theory, it quantifies the amount of "information" or "uncertainty" in a node.

3.  Classification Error (less common)

Measures the frequency of the most common class.

4. Variance (for Regression Trees)

Used instead of Gini or Entropy when the target is numeric.

Que8. What is Pre-Pruning in Decision Trees ?

Ans: In pre-pruning, the decision tree stops splitting a node if a certain condition is met before reaching a pure leaf (i.e., before all samples in a node belong to the same class).

This contrasts with post-pruning, where the full tree is built first and then pruned back.

âœ… Benefits of Pre-Pruning

Prevents overfitting to noisy or small datasets

Reduces model complexity, making the tree easier to interpret

Speeds up training time





Que9. What is Post-Pruning in Decision Trees ?

Ans: Post-pruning, also known as cost-complexity pruning or reduced-error pruning, is a technique used after a full decision tree is built to simplify the model and reduce overfitting by removing sections of the tree that provide little or no predictive power.

ðŸ” How It Works

Grow the full tree until all leaves are pure (or other stopping criteria are met).

Evaluate subtrees or branches from the bottom up.

Replace a subtree with a leaf node if pruning improves performance on a validation set or does not significantly degrade it.

Repeat until no further improvements can be made.

Que13. How does a Decision Tree handle missing values ?

Ans: A Decision Tree can handle missing values in several intelligent ways, depending on the implementation and algorithm. Hereâ€™s a breakdown of how missing values are managed:

ðŸ§© 1. Skipping the Feature (Simple Strategy)

If a feature has missing values, the tree can ignore that feature for splitting at a given node.

This is the simplest approach, but may reduce model accuracy.

ðŸ”€ 2. Surrogate Splits (Advanced Strategy)

Used in CART and tools like R's rpart or XGBoost.

When a value is missing for the primary split feature, the model uses a surrogate feature â€” another feature that provides a similar split behavior.

How it works:

During training, the algorithm identifies alternate features (surrogates) that mimic the behavior of the main split.

At prediction time, if a value is missing, the best surrogate is used instead.

âœ… Advantage: Doesn't discard data with missing values.

âš–ï¸ 3. Assigning Probabilities or Weights
Some libraries (e.g., XGBoost, LightGBM) learn default directions when missing values occur (e.g., "go left if missing").

Training samples with missing values are sent in both directions with fractional weights, and the best split is learned.

âœ… Used for handling large-scale datasets with missing values effectively.

ðŸ§¹ 4. Imputation Before Training

A preprocessing step where missing values are filled in using:

Mean/median/mode

k-NN

Regression

Forward/Backward fill (time-series)

âœ… Simple but may introduce bias if the imputation is not representative.

Que14. How does a Decision Tree handle categorical features ?

Ans:  How Decision Trees Handle Categorical Features

âœ… 1. Native Handling (in ID3, C4.5, LightGBM, CatBoost)

Some algorithms can directly split on categorical values without encoding. For example:

For feature Color with values {Red, Green, Blue}, the tree might split like:

IF Color = Red â†’ go left, ELSE â†’ go right

Advanced implementations (e.g., LightGBM, CatBoost) group categories based on statistics (e.g., mean target value) for better performance.

ðŸ§  2. Splitting Criteria

For binary classification with categorical features:

All possible subsets of categories can be evaluated to find the best split.

Example for Color = {Red, Green, Blue}:

Possible splits:

{Red} vs {Green, Blue}

{Green} vs {Red, Blue}

{Red, Green} vs {Blue}

The tree evaluates each split using Gini impurity or Entropy and chooses the best.

ðŸ› ï¸ 3. With Encoding (in scikit-learn and similar libraries)

Since scikit-learn's Decision Trees donâ€™t natively support strings, categorical features must be numerically encoded:

A. Label Encoding
Converts categories into integers (e.g., Red = 0, Green = 1, Blue = 2)

âš ï¸ Can introduce misleading ordinal relationships

B. One-Hot Encoding
Creates binary columns for each category:
The tree treats each one-hot feature separately.

C. Ordinal Encoding (if order matters)
E.g., Size = {Small = 0, Medium = 1, Large = 2}
